```python
from langchain.embeddings.base import Embeddings
import ollama

class OllamaEmbeddings(Embeddings):
    def __init__(self, model_name="all-minilm"):
        self.model_name = model_name

    def embed_documents(self, texts):
        # Ensure it's a list
        if isinstance(texts, str):
            texts = [texts]
        resp = ollama.embed(model=self.model_name, input=texts)
        return resp.embeddings  # list of vectors

    def embed_query(self, text):
        # For single query
        resp = ollama.embed(model=self.model_name, input=[text])
        return resp.embeddings[0]  # single vector
```

```python
from langchain_community.vectorstores import Chroma

# Your documents
texts = [doc.page_content for doc in text_document_2]

# Create Ollama embedding object
embeddings = OllamaEmbeddings(model_name="all-minilm")

# Create Chroma vectorstore
db = Chroma.from_texts(texts, embedding=embeddings)
```

```python
# vector database

query = "who are the authors of attention all you need research paper?"
result = db.similarity_search(query)
```
